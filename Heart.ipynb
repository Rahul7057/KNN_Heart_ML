```python
x = df.drop('Status_Dead',axis=1).values
x
```

```python
y = df['Status_Dead'].values
y
```

```python
np.sqrt(5057)
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=71,algorithm='brute',weights='distance')
```

```python
model.fit(xtrain,ytrain)
```

```python
ypred=model.predict(xtest)
ypred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
acc=accuracy_score(ypred,ytest)
acc
```

```python
cm=confusion_matrix(ypred,ytest)
cm
```

```python
model_cv=KNeighborsClassifier(n_neighbors=71,algorithm='kd_tree',weights='distance')
cv=cross_val_score(model,x,y,cv=15,scoring='accuracy')
cv.mean()
```
